```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar os dados (assumindo que estão em um arquivo CSV)
# df = pd.read_csv('solicitacoes_emprestimo.csv')
# Como não temos o arquivo real, vou criar um dataframe de exemplo para ilustrar
# Na prática, você usaria o arquivo real fornecido

# Criando dados de exemplo (substitua pelo seu dataset real)
data = {
    'Salario_Anual': [50000, 80000, 45000, 120000, 75000, 90000, 60000, 30000, 95000, 110000],
    'Dividas': [20000, 15000, 30000, 5000, 25000, 10000, 18000, 35000, 12000, 8000],
    'Score': [0.85, 0.92, 0.78, 0.95, 0.88, 0.91, 0.82, 0.65, 0.93, 0.94],
    'Idade': [35, 42, 28, 50, 38, 45, 32, 25, 48, 52],
    'Resultado': [2, 3, 1, 3, 2, 3, 2, 1, 3, 3]
}

df = pd.DataFrame(data)

# 1. Análise exploratória e pré-processamento
print(df.head())
print(df.describe())

# Verificar valores nulos
print(df.isnull().sum())

# Visualizar distribuição das variáveis
df.hist(figsize=(10,8))
plt.tight_layout()
plt.show()

# Verificar correlações
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Matriz de Correlação')
plt.show()

# 2. Tratamento de outliers (opcional - pode ser ajustado conforme necessidade)
# Vamos remover outliers extremos nas variáveis numéricas
def remove_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

df_clean = remove_outliers(df, ['Salario_Anual', 'Dividas', 'Score', 'Idade'])

# 3. Preparação dos dados
X = df_clean.drop('Resultado', axis=1)
y = df_clean['Resultado']

# Dividir em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Normalizar os dados
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 4. Treinamento do modelo KNN
# Encontrar o melhor valor de K
error_rates = []
# Adjust the range of k to not exceed the number of samples in the training set
for k in range(1, min(11, len(X_train_scaled) + 1)):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    preds = knn.predict(X_test_scaled)
    error_rates.append(np.mean(preds != y_test))

plt.figure(figsize=(10,6))
plt.plot(range(1, min(11, len(X_train_scaled) + 1)), error_rates, marker='o')
plt.xlabel('Valor de K')
plt.ylabel('Taxa de Erro')
plt.title('Taxa de Erro vs. Valor de K')
plt.show()

# Escolher o K com menor taxa de erro
best_k = error_rates.index(min(error_rates)) + 1
print(f"Melhor valor de K: {best_k}")

# Treinar modelo com o melhor K
knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(X_train_scaled, y_train)

# 5. Avaliação do modelo
y_pred = knn.predict(X_test_scaled)

print("\nRelatório de Classificação:")
print(classification_report(y_test, y_pred))

print("\nMatriz de Confusão:")
print(confusion_matrix(y_test, y_pred))

# 6. Exemplo de previsão
def predict_loan_status(salario, dividas, score, idade):
    input_data = np.array([[salario, dividas, score, idade]])
    input_scaled = scaler.transform(input_data)
    prediction = knn.predict(input_scaled)[0]

    if prediction == 1:
        return "Não elegível - Solicitação recusada/negada"
    elif prediction == 2:
        return "Elegível com análise - Necessita análise detalhada"
    else:
        return "Elegível - Solicitação atendida imediatamente"

# Testar a função de previsão
print("\nExemplo de previsão:")
print(predict_loan_status(65000, 20000, 0.87, 40))
```
